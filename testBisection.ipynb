```
import numpy as np
import matplotlib.pyplot as plt

## define function ####
def f1(x):
    return x**3 + 10*x - 7

# plot the graph of function
xvalues = np.linspace(-5,5,100)

plt.figure()
plt.plot(xvalues,f1(xvalues), "r:")
plt.title("Plot of function f(x) = x^3 + 10x - 7")
plt.xlabel('x')
plt.grid()
plt.show()
```

```
def BisectionVec(a,b,TOL,f):
    
    print("************Bisection Method ************")
    Error = TOL + 100
    xvec  = np.array([]) 
    ErrorVec = np.array([])
    NumIter = 0
    while Error > TOL:

        x = (a + b)/2  # Approximate value from Bisection

        Error = abs(f(x)) # Residual

        # record x values and Errors
        xvec = np.concatenate((xvec, x), axis = None)
        ErrorVec = np.concatenate((ErrorVec, Error), axis = None)
        NumIter = NumIter + 1   

        # print out 
        print( "Iter:%d, [a,b] = [%0.3f, %0.3f], x = %0.4f, residual |f(x)| = %0.4f" %( NumIter, a, b, x, Error))

        if f(a)*f(x) < 0:
            b = x 
        else:
            a = x 

         
        # End of "while" loop

    print(" The final approximation is ", x)

    return x, xvec, ErrorVec, NumIter

### Call function BisectionVec

a = 0
b = 2
TOL = 0.0001
f = f1
x, xvec, ResVec, NumIterAll = BisectionVec(a,b,TOL,f)

# Plot of solution
IterVec = np.arange(1, NumIterAll +1, 1) # [1,2,3,..., NumIterAll]
plt.figure,
plt.plot(IterVec, xvec, "r*--")
plt.title("Approximate root of f(x) = 0 from Bisection Method")
plt.xlabel("Iteration Number")
plt.ylabel("Approximate Value")
plt.grid()
plt.show()
# Plot of residual
plt.figure,
plt.plot(IterVec, ResVec, "go--")
plt.title(" Residual |f(x)| from Bisection Method")
plt.xlabel("Iteration Number")
plt.ylabel("Residual")
plt.grid()
plt.show()

# Plot of residual (log scale)
plt.figure,
plt.semilogy(IterVec, ResVec, "gs--")
plt.title(" Residual |f(x)| from Bisection Method")
plt.xlabel("Iteration Number")
plt.ylabel("Residual")
plt.grid()
plt.show()
```
