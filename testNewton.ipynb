```
import numpy as np
import matplotlib.pyplot as plt

## define function ####
def f1(x):
    return x**3 + 10*x - 7

## define derivative of function ###

def f1diff(x):
    return 3*(x**2) + 10
    

# plot the graph of function
xvalues = np.linspace(-5,5,100)

plt.figure()
plt.plot(xvalues,f1(xvalues), "r:")
plt.title("Plot of function f(x) = x^3 + 10x - 7")
plt.xlabel('x')
plt.grid()
plt.show()
```

```
## function for Newton's method

def NewtonVec(x0, TOL, f, fdiff):


    x = x0
    Error = abs(f(x)) # intial residual
    xvec = np.array([x])
    ErrorVec = np.array([Error])
    NumIter  = 0
    
    print("****Newton's method ******")
    print("Iter:%d, x = %0.4f, residual |f(x)| = %0.4f" %(NumIter, x, Error) )

    while Error > TOL:
          
          x  = x - f(x)/fdiff(x) # approximation from Newton's method
          Error = abs(f(x))  # residual

          NumIter = NumIter  + 1

          # Print out info
          print("Iter:%d, x = %0.4f, residual |f(x)| = %0.4f" %(NumIter, x, Error) )

          # Keep the approximations and errors
          xvec = np.concatenate( (xvec, x), axis = None)
          ErrorVec = np.concatenate( (ErrorVec, Error ), axis = None)


          ## end of while loop
    print("The final approximate root is ", x)
    return x, xvec, ErrorVec, NumIter

## Call NewtonVec function
x0 = 0
TOL = 0.0001
 
x, xvec, ResVec, NumIterAll = NewtonVec(x0, TOL, f1, f1diff)    
IterVec = np.arange(0, NumIterAll +1, 1) # [0,2,..., NumIterAll]
# Plot of approximate root from Newton's
plt.figure()
plt.plot(IterVec, xvec, "r:*")
plt.title("Approximate root from Newton's method")
plt.xlabel("Iteration Number")
plt.ylabel("Approximate value")
plt.show()
```

```
# Plot of residual from Newton's
plt.figure()
plt.semilogy(IterVec, ResVec, "g:*")
plt.title(" Residual from Newton's method (log scale)")
plt.xlabel("Iteration Number")
plt.ylabel("Residual")
plt.grid()
plt.show()
```
